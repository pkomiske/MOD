## nPV Reweighting

```python
cms = mod.MODDataset(utils.path('cms', 'h5', 'CMS_Jet300_pT375-infGeV'), 'abs_jet_eta < 1.9 & quality >= 2',
                     store_pfcs=False)
sim = mod.MODDataset(utils.path('sim', 'h5', 'SIM_Jet300_pT375-infGeV_noparticles'), 
                     'abs_jet_eta < 1.9 & quality >= 2', store_pfcs=False, store_gens=False)
```

```python
npvs = {}
weights = {}
for name,dset in zip(['cms', 'sim'], [cms, sim]):
    rnlbevns = set()
    npvs_local = []
    weights_local = []
    for t in zip(dset.rns, dset.lbns, dset.evns, dset.npvs, dset.weights):
        if t[:3] not in rnlbevns:
            rnlbevns.add(t[:3])
            npvs_local.append(t[3])
            weights_local.append(t[4])
    npvs[name] = np.asarray(npvs_local)
    weights[name] = np.asarray(weights_local)
```

```python
bins = np.arange(np.max(np.concatenate((npvs['cms'], npvs['sim']))) + 2)
cms_npv_hist = np.histogram(npvs['cms'], bins=bins, weights=weights['cms'], density=True)[0]
sim_npv_hist = np.histogram(npvs['sim'], bins=bins, weights=weights['sim'], density=True)[0]
```

```python
npv_hist_ratios = cms_npv_hist/(sim_npv_hist + 10**-30)
```

## Residual Normalization Correction

```python
bins = np.linspace(375, 2250, 76)
cms_pt_hist = np.histogram(cms.corr_jet_pts, bins=bins, weights=cms.weights)

sim_kfactors = np.interp(sim.corr_jet_pts, data[:,0], data[:,1])
sim_npv_weights = npv_hist_ratios[np.asarray(sim.npvs, dtype=int)]
sim_pt_hist = np.histogram(sim.corr_jet_pts, bins=bins, weights=sim.weights*sim_kfactors*sim_npv_weights)
```

```python
plt.plot(bins[:-1], cms_pt_hist[0], 'o-', color='black')
plt.plot(bins[:-1], sim_pt_hist[0], 'o-', color='orange')
plt.xlim(0, np.max(bins))
plt.yscale('log')
plt.show()
```

```python
residual_factor = cms_pt_hist[0][0]/sim_pt_hist[0][0]
```
